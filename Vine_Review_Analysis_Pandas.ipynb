```python
import pandas as pd
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
vine_review = "vine-table.csv"

vine_review_df = pd.read_csv(vine_review)
vine_review_df.head(5)

```

```python
vine_review_df.tail(5)
```

```python
vine_df = vine_review_df.loc[(vine_review_df ["total_votes"] >= 20)]
vine_df.head(10)
```

```python
vine_votes_df = vine_df.loc[((vine_df ["helpful_votes"] / vine_df["total_votes"]) >= 0.5)]
vine_votes_df.head(10)
```

```python

vine_only_df = vine_votes_df.loc[(vine_votes_df['vine'] == 'Y')]
vine_only_df.head(10)
```

```python
nonVine_df = vine_votes_df.loc[(vine_votes_df['vine'] == 'N')]
nonVine_df.head(10)
```

```python
total_paid_review = vine_only_df["star_rating"].count()
total_paid_review
```

```python
total_unpaid_review = nonVine_df["star_rating"].count()
total_unpaid_review
```

```python
total_paid_five_star_review = vine_only_df.loc[(vine_only_df["star_rating"] == 5 ), "star_rating"].count()
#total_5_star_review = vine_only_df.filter(vine_only_df.star_rating == 5).count()
total_paid_five_star_review
```

```python
total_unpaid_five_star_review = nonVine_df.loc[(nonVine_df["star_rating"] == 5 ), "star_rating"].count()
#total_5_star_review = vine_only_df.filter(vine_only_df.star_rating == 5).count()
total_unpaid_five_star_review

```

```python
paid_five_star_review_percentage = (total_paid_five_star_review / total_paid_review) * 100
paid_five_star_review_percentage
```

```python
non_paid_five_star_review_percentage = (total_unpaid_five_star_review / total_unpaid_review) * 100
non_paid_five_star_review_percentage
```
